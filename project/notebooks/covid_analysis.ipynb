## Column type definition

```python
df_train.info()
```

```python
df_test.info()
```

The column "Location" is type object, changing it to type category. Also, the columns "Weekly Cases", "Weekly Deaths", "Total Vaccinations", "People Vaccinated", "People Fully Vaccinated", "Total Boosters", "Daily Vaccinations", "Daily People Vaccinated" should be type int but we cannot convert it since there are null values on them.

```python
df_train["Location"] = df_train["Location"].astype('category')
df_train.info()
```

```python
df_test["Location"] = df_test["Location"].astype('category')
df_test.info()
```

## Treating missing data

```python
df_train.loc[df_train["Weekly Deaths"].isna()].info()
```

```python
df_test.loc[df_test["Weekly Deaths"].isna()].info()
```

Removing occurences of null values on columns Weekly Deaths and Weekly Deaths per Million.

```python
df_train.dropna(subset=["Weekly Deaths", "Weekly Deaths per Million"], inplace=True)
df_test.dropna(subset=["Weekly Deaths", "Weekly Deaths per Million"], inplace=True)
```

```python
df_train.info()
```

```python
df_train.loc[df_train["Weekly Cases"].isna()].info()
```

```python
df_test.loc[df_test["Weekly Cases"].isna()].info()
```

Removing occurences of null values on columns Weekly Cases and Weekly Cases per Million.

```python
df_train.dropna(subset=["Weekly Cases", "Weekly Cases per Million"], inplace=True)
df_test.dropna(subset=["Weekly Cases", "Weekly Cases per Million"], inplace=True)
```

```python
df_train.info()
```

## Normalization and discretization

Creating a new feature related to the vaccine existence.
